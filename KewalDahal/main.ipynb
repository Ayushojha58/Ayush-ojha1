```python
#project 1
import os
from datetime import datetime 

def create_journal():
    today = datetime.now().strftime('%Y-%m-%d')
    print(today)
    fileName = (f"{today}.txt")

    #asking user to enter their journal for the day.
    print("\nHow was your day? Write below:")
    journalEntry = input(">start writing your journal mate> ")


    with open(fileName, 'w') as file :
        file.write(journalEntry)

    print(f"\n your journal for the date of {today} has been recorded")

def read_journal():
    # Get all files ending in .txt
    files = [f for f in os.listdir() if f.endswith('.txt')]

    if not files:
        print("No journal entries found.")
        return

    # Display available entries
    print("Available journal entries:")
    for file in files:
        print(file)

    # Prompt user to select a file
    selected_file = input("Enter the date of the journal entry you want to read (YYYY-MM-DD): ") + ".txt"
    
    if selected_file not in files:
        print("File not found!")
        return

    # Read and display the selected journal entry
    with open(selected_file, 'r') as file:
        print(f"\nJournal entry for {selected_file[:-4]}:\n")
        print(file.read())

def main():
    while True:
        print("\nJournal Program")
        print("1. Write a new journal entry")
        print("2. Read a previous journal entry")
        print("3. Exit")
        
        choice = input("\nWhat would you like to do? (1/2/3): ")
        
        if choice == '1':
            create_journal()
        elif choice == '2':
            read_journal()
        elif choice == '3':
            print("\nGoodbye! Take care!")
            break
        else:
            print("\nInvalid choice. Please enter 1, 2, or 3.")

if __name__ == "__main__":
    main()

```

```python
import os
from datetime import datetime, timedelta

def record_weather_data():
    today = datetime.now().strftime('%Y-%m-%d')
    filename = f"{today}.txt"
    
    #using empty list to jold the data provided by user
    weather_data = []

    # Taking weather data from the user
    print("Enter today's weather details:")
    temperature = input("Temperature in °C: ")
    humidity = input("Humidity in %: ")
    conditions = input("Conditions: ")
    weather_data.append((today, temperature, humidity, conditions))

    with open(filename, 'w') as file:
        for detail in weather_data:
            file.write(f"Date: {detail[0]} \n")
            file.write(f"Temperature: {detail[1]}°C \n")
            file.write(f"Humidity: {detail[2]}% \n")
            file.write(f"Conditions: {detail[3]} \n")

    print(f"weather data for {today} has been recorded in {filename}.")

def view_weather_data():
    files = [f for f in os.listdir() if f.endswith('.txt')]
    
    if not files:
        print("No weather data files found.")
        return
    

    print("Available weather data files:")
    for file in files:
        print(file)


    selected_date = input("Enter the date of the weather data you want to read (YYYY-MM-DD): ") + ".txt"
    
    if selected_date not in files:
        print("File not found!")
        return
    
    with open(selected_date, 'r') as file:
        print(f"\nWeather data for {selected_date[:-4]}:\n") #-4 is used for trimming the '.txt' from the filename(took help from claude)
        print(file.read())

def get_last_week_dates():
    today = datetime.now()
    return [(today - timedelta(days=i)).strftime('%Y-%m-%d') for i in range(7)]

def readTemp_fromFile(filename):
    try:
        with open(filename, 'r') as file:
            for line in file:
                if line.startswith("Temperature:"):
                    temp_str = line[13:].strip()  # skip the first 13 characters
                    temp_str = temp_str.replace("°C", "")
                    return float(temp_str)
    except FileNotFoundError:
        print(f"File {filename} not found.")
        return None

def display_trends():
    lastWeek_dates = get_last_week_dates()

    temperatures = []
    print("Calculating temperature trends for the last week...")
    for date in lastWeek_dates:
        filename = f"{date}.txt"
        temp = readTemp_fromFile(filename)
        if temp is not None:
            temperatures.append(temp)
    
    if temperatures:
        average = sum(temperatures) / len(temperatures)
        heighest = max(temperatures)
        lowest = min(temperatures)
        print(f"\nAverage temperature over the last week: {average:.2f}°C")
        print(f"Highest temperature: {heighest:.2f}°C")
        print(f"Lowest temperature: {lowest:.2f}°C")
    else:
        print("No temperature data available for the last week.")

def main():
    while True:
        print("\nWeather Logger Menu")
        print("1. Record today's weather data")
        print("2. View weather data")
        print("3. Display temperature trends for the last week")
        print("4. Exit")

        choice = input("Enter your choice (1-4): ")

        if choice == '1':
            record_weather_data()
        elif choice == '2':
            view_weather_data()
        elif choice == '3':
            display_trends()
        elif choice == '4':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

if __name__ == "__main__":
    main()
```
